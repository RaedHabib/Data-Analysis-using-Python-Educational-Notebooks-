Sometimes we might face a problem with our data that some feature which suppose to represent numerical data is recognized by Python as object (string) like the **"households"** feature in our case, and that means that there are some elements (values) in this column or feature which are not numerical values, so we will have to fix this issue because these values are misleading and are considered as hidden missing values; because we won't be able to find them directly by checking for missing values (with isnull() method).

So, the first thing to do is to locate those values:

```python
house['households'].value_counts()
```

Now we can see that someone has put the word "no" instead of putting or writting the correct numerical value, so, we will have to change those values back to 'NAN' for Python to be able to identify them as missing values then deal with them properly (eliminate them) like we did in the above examples:

```python
house['households'] = house['households'].replace("no" , np.nan)
# Replacing the string values "no" with the nonetype values "NAN" to be recognized as missing values
```

```python
house.info()
```

We notice that the previuos action didn't change the datatype (object or string) of the column "households", so we will have to change it back to integer (because it represents people) manually.

```python
house.isnull().sum() 
# We will notice that the recognized missing values of "households" has increased due to revealing all 
# the hidden missing values in the shape of string objects.
```

Python Pandas fillna() method has a very useful feature which is a parameter called "method" in which we can pass an argument called "bfill" or "backfill" (both are correct) and it will automatically fill any gaps (missing values) in any column or feature with the directly next non NAN value to that gap. 

```python
data['tenure'].fillna(method = "bfill") # It's a very useful technique.
```

But it has a great disadvatage in case of more than NAN values located after each other in a sequence, so it will only fill the last one. So we will still have missing values.

```python
data.isnull().sum()
```

As we can see there are still missing values in the "tenure" feature. Let's try something else:

```python
fill_list = {"SeniorCitizen" : int(data['SeniorCitizen'].mode()) , 
            "tenure" : data['tenure'].mean()}
# As we noticed we used the mode filling with discrete values (SeniorCitizen has values of 0 and 1 only)
# SeniorCitizen can be considered in this example to be categorical as it only has 0 and 1 values which
# represents True or False, but anyways it doesn't matter to us in this case; as we can use the mode method
# with discrete or categorical values all the same.

# We also noticed the use of the int() function, because the mode() function returns values of float type.



# This is an effective and quick way of replacing (filling) missing values for more than one feature at
# the same time.
```

There's another way of changing the data type of the passed argument like we did with the int() function in the last code, which is a Pandas method called: pandas.to_numeric(). For more info refer to this page:
https://www.geeksforgeeks.org/python-pandas-to_numeric-method/

```python
data.fillna(fill_list , inplace = True)
```

##  Data Binning 
Data binning, or bucketing is a data pre-processing method used to minimize the effects of small observation errors. The original data values are divided into small intervals known as bins and then they are replaced by a general value calculated for that bin. This has a smoothing effect on the input data and may also reduce the chances of overfitting in the case of small datasets.

It's simply like categorizing values in a specific feature into some categories and represents them in a new (added) feature in our DataFrame. We will be needing this technique later on in the near future.

For example, in our Data what if we want to reward our customers with offers specified for each category of them based on their service usage?

```python
def offer(row):
    if row['gender'] == "Male" and row['SeniorCitizen'] == 1:
        return "net"
    elif row['gender'] == "Male" and row['SeniorCitizen'] == 0:
        return "coins"
    elif row['gender'] == "Female" and row['SeniorCitizen'] == 0:
        return "net_and_coins"
    elif row['gender'] == "Female" and row['SeniorCitizen'] == 1:
        return "net_flex"
    else:
        return "no_offer"
```

```python
data['offer'] = data.apply(offer , axis = 1)
# Here we are applying the offer function to each row in the DataFrame for each specified column values.
# For further explanation:
# We are iterating through the columns "SeniorCitizen" and "gender" for each row in our DataFrame
```
