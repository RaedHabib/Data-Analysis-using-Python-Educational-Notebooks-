##  Data Binning 
Data binning, or bucketing -- groups data in bins (or buckets) -- is a data pre-processing method used to minimize the effects of small observation errors. The original data values are divided into small intervals known as bins and then they are replaced by a general value calculated for that bin. This has a smoothing effect on the input data and may also reduce the chances of overfitting in the case of small datasets.

It's simply like categorizing values in a specific feature into some categories and represents them in a new (added) feature in our DataFrame. We will be needing this technique later on in the near future.

Binning can be applied to convert numeric values to categorical or to sample (quantise) numeric values.

Binning is a technique for data smoothing. Data smoothing is employed to remove noise from data.There are three techniques for data smoothing:
>Binning

>Regression (to be discussed later)

>Outlier analysis (also to be discussed later)

For example, in our Data what if we want to reward our customers with offers specified for each category of them based on their service usage?

```python
def offer(row):
    if row['gender'] == "Male" and row['SeniorCitizen'] == 1:
        return "net"
    elif row['gender'] == "Male" and row['SeniorCitizen'] == 0:
        return "coins"
    elif row['gender'] == "Female" and row['SeniorCitizen'] == 0:
        return "net_and_coins"
    elif row['gender'] == "Female" and row['SeniorCitizen'] == 1:
        return "net_flex"
    else:
        return "no_offer"
```

```python
data['offer'] = data.apply(offer , axis = 1)
# Here we are applying the offer function to each row in the DataFrame for each specified column values.
# For further explanation:
# We are iterating through the columns "SeniorCitizen" and "gender" for each row in our DataFrame
```

 > DataFrame : when DataFrame.agg is called with several functions

## Encoding
It is the process of converting data from one form to another.

There are many types of different encodings but we are going to focus only on:


â‘ Label Encoding: It is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering (starting from zero), so we are converting these labels into a numeric form which is the machine-readable form.

It's usually used for ordinal (ranked) data.

Note that: sometimes label encoders don't bother with the alphabetical ordering, and just assign the integers based on the variables order in the Dataset as they are without rearranging them.
<br>
    
â‘ One-hot encoding: It is essentially the representation of categorical variables as binary vectors. These categorical values are first mapped to integer values (after splitting them to categories based on the unique values). Each integer value is then represented as a binary vector that is all 0s (except the index of the integer which is marked as 1)

It's usually used for nominal data.

<br>
Most of the time we prefer to use the label encoder specifically when it comes to modeling and making predictions; because one-hot encoder increases the dimension (no. of features), while we will need to reduce the dimentionality of our data, so label encoder would be prefered, but that doesn't neglect the importance of one-hot encoder when it comes to doing analysis on our data.

Instead, fit_transform() is used to get both works done.

<br>
<br>

> transform(): For changing the data, we probably do transform in the transform() method, where we apply the calculations that we have calculated in fit() to every data point in the feature.

## Very Important Point (VIP)
The whole time we've been calling columns in our Datasets (DataFrames) as "features", but this is not 100% accurate; as "features" refers only to input data (columns).
> In every Dataset we have multiple input data (independent variables) and one single target data (dependent variable) that we are going to predict its values based on our input data (columns) or features


<br>

To be more accurate "LabelEncoder()" is originally designed to encode only the target data (column) because it takes only one column at a time, **while in order to encode the features of our Dataset (input columns) we use another encoding type designed specifically to this task and can take multi-columns at a time, and it's called: "Ordinal Encoding" ---> "OrdinalEncoder()"**

**It has the same functionality as the "LabelEncoder()", but OrdinalEncoder can fit data that has the shape of (n_samples, n_features) {2D data or array} while LabelEncoder can only fit data that has the shape of (n_samples,) {1D data or array} (though in the past one used LabelEncoder within the loop to handle what has been becoming the job of OrdinalEncoder now).**

In order to understand what is "Compressed Sparse Row", let's first introduce a very important terminology, they are actually two related (opposite) terminologies:

* Sparse vs Dense formats:
Matrices that contain mostly zero values are called **"Sparse"**, while matrices where most of their values are non-zero, called **"Dense"**


* For "Sparce Matrix" format there are two compressed formats which we use to save memory:

1- Compressed Sparse Row (CSR) or the Yale Format 
> values are stored row by row. 

2- Compressed Sparse Column (CSC)
> values are stored column by column.

Mainly, they are used for write-once-read-many tasks (like the one-hot encoding task we did above). They are basically a representation of the sparse matrix but in 3 (1D) arrays, one for storing the nonzero values, the other two for their (each nonzero value) respective row and column indices.


**Without wasting time on diving in unnecessary details (at least for now), let's just know these concepts as they are for now and understand their main characteristics as stated above.**

## Scaling
Feature Scaling is the process of normalising the range of features in a dataset. Real-world datasets often contain features that are varying in degrees of magnitude, range and units. Therefore, in order for machine learning models to interpret these features on the same scale, we need to perform feature Scaling

Methods for Scaling
Now, since you have an idea of what is feature scaling. Let us explore what methods are available for doing feature scaling. Of all the methods available, the most common ones are:


**1- Normalization**

Also known as min-max scaling or min-max normalization, it is the simplest method and consists of rescaling the range of features (real-valued numeric values) to the range in [0, 1]. The general formula for normalization is given as:

Here, max(x) and min(x) are the maximum and the minimum values of the feature respectively.

Data normalization is used in machine learning to make model training less sensitive to the scale of features, which allows our model to be more accurate.

This might be useful in somecases where all parameters need to have the same positive scale.However, impact of Outliers is very high in Normalization.

* examples of usage:
> - Normalization is good to use when the distribution of data does not follow a Gaussian distribution. It can be useful in algorithms that do not assume any distribution of the data like K-Nearest Neighbors.
> <br>  <br>
> - In Neural Networks algorithm that require data on a 0â€“1 scale, normalization is an essential pre-processing step. Another popular example of data normalization is image processing, where pixel intensities have to be normalized to fit within a certain range (i.e., 0 to 255 for the RGB color range).

--------------------

**2- Standardization**

Feature standardization makes the values of each feature in the data have zero mean (mean (Î¼) of 0) and standard deviation (Ïƒ) of 1 (unit variance).. The general method of calculation is to determine the distribution mean and standard deviation for each feature and calculate the new data point by the following formula:

As we can see from the formula that it's that of the Z-score, and that's why Standardization is often called Z-score Normalization; and hence, the most of the values (â‰ˆ100%) will be in the range [-3:+3] (since Ïƒ = 1).


For most applications standardization is recommended.

* examples of usage:
> - Standardization can be helpful in cases where the data follows a Gaussian distribution.Though this does not have to be necessarily true. Since standardization does not have a bounding range, so, even if there are outliers in the data, they will not be affected by standardization.
> <br> <br>
> - In clustering analyses, standardization comes in handy to compare similarities between features based on certain distance measures.
> <br> <br>
> - Another prominent example is the Principal Component Analysis, where we usually prefer standardization over Min-Max scaling since we are interested in the components that maximize the variance.

In the presence of outliers, StandardScaler does not guarantee balanced feature scales, due to the influence of the outliers while computing the mean and standard deviation. This leads to the shrinkage in the range of the feature values. 


To conclude, we can always start by fitting our model to raw, normalized, and standardized data and compare the performance for the best results.

<br>

### Very Important Note 
It's highly recommended to use only one method of the two scaling methods above with our data, either normalize or standardize our features we can not split the data into two groups with different scaling (we can not do both with the same data!).

Due to the impact of outliers on both min-max scaling and standard scaling, by using RobustScaler(), we can remove the outliers and then use either StandardScaler() or MinMaxScaler() for preprocessing the dataset.

Scale features using statistics that are robust (immune) to outliers.

This Scaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).
